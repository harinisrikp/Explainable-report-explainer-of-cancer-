```python
import pandas as pd

# Load the dataset
file_path = "D:\mini 2\kag_risk_factors_cervical_cancer.csv"
data = pd.read_csv(file_path)

# Display basic information
print(data.info())
print(data.head())
```

```python
#print missing value
print(data.isnull().sum())
```

```python
data = data.drop(columns=['STDs: Time since last diagnosis'])
```

```python
data = data.drop(columns=['STDs: Time since first diagnosis'])
```

```python
#print missing value
print(data.isnull().sum())
```

```python
import numpy as np

# Replace non-numeric values (e.g., '?') with NaN
data[numerical_cols] = data[numerical_cols].replace('?', np.nan)

# Convert columns to numeric
for col in numerical_cols:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Fill missing values with the median
for col in numerical_cols:
    data[col] = data[col].fillna(data[col].median())
```

```python
import pandas as pd

# Fill missing numerical values with the median
numerical_cols = ['Number of sexual partners', 'First sexual intercourse', 'Num of pregnancies', 
                  'Smokes (years)', 'Smokes (packs/year)', 'Hormonal Contraceptives (years)', 
                  'IUD (years)', 'STDs (number)']
for col in numerical_cols:
    data[col] = data[col].fillna(data[col].median())

# Fill missing categorical values with 0 (assuming "No"/"0" indicates absence)
categorical_cols = ['Smokes', 'Hormonal Contraceptives', 'IUD', 'STDs', 'STDs:condylomatosis',
                    'STDs:cervical condylomatosis', 'STDs:vaginal condylomatosis',
                    'STDs:vulvo-perineal condylomatosis', 'STDs:syphilis', 'STDs:pelvic inflammatory disease',
                    'STDs:genital herpes', 'STDs:molluscum contagiosum', 'STDs:AIDS', 'STDs:HIV',
                    'STDs:Hepatitis B', 'STDs:HPV']
for col in categorical_cols:
    data[col] = data[col].fillna(0)

# Verify if missing values are handled
print(data.isnull().sum())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Step 1: Define the feature matrix (X) and target variable (y)
X = data.drop('Biopsy', axis=1)  # Replace 'Biopsy' with your target column name
y = data['Biopsy']  # Replace 'Biopsy' with your target column name

# Step 2: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Step 3: Initialize and train the Random Forest classifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train, y_train)

# Step 4: Make predictions
y_pred = clf.predict(X_test)

# Step 5: Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
